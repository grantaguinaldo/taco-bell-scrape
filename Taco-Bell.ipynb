```python
import requests as r
from bs4 import BeautifulSoup as soup
```

```python
url = 'https://www.yelp.com/biz/taco-bell-los-angeles-24'
```

```python
data = r.get(url)
```

```python
data
```

```python
page_data = soup(data.text, 'html.parser')
```

```python
page_data
```

```python
print(page_data.prettify())
```

```python
page_data.find_all('div', {'class': 'review-content'})[0]
```

```python
soup_data = page_data.find_all('div', {'class': 'review-content'})
```

```python
soup_data[0].find('p').text
```

```python
soup_data[0].find('img')['alt'].split(' ')[0]
```

```python
soup_data[0].find('img')['alt']
```

```python
#Loop over soup_data
#Store the items that we found, in our survey in a DataFrame (text, full star rating, num star rating)
#Send the DataFrame to a CSV so we can store it. 
```

```python
import pandas as pd
df = pd.DataFrame()
```

```python
for each in soup_data:
    each_review = each.find('p').text
    num_stars = each.find('img')['alt'].split(' ')[0]
    full_stars = each.find('img')['alt']
    
    df = df.append(pd.DataFrame({'review': each_review, 'star_number': num_stars, 'full_rating': full_stars}, index = [0]), ignore_index = True)
```
